```python
# Generate a function 
def generate_probability_matrix(n=countmatrix):

# make a copy of the count matrix
 countmatrix.copy()
# Add the copy to a variable 
probability_matrix = countmatrix.copy()
# Define the value of each cell
value1 = probability_matrix.iloc[0,0]
value2 = probability_matrix.iloc [1,0]
value3 = probability_matrix.iloc [2,0]
value4 = probability_matrix.iloc [0,1] 
value5 = probability_matrix.iloc [1,1]
value6 = probability_matrix.iloc [2,1]
value7 = probability_matrix.iloc [0,2] 
value8 = probability_matrix.iloc [1,2] 
value9 = probability_matrix.iloc[len(probability_matrix.index) - 1,len(probability_matrix) - 1]
# Define the total number of patients
totalnoofpatients = countmatrix.iloc[len(countmatrix.index) - 1,len(countmatrix) - 1]
# Create a list for the newvalues
probvalues = []
# Create a list of the values
values = [value1, value2, value3, value4, value5, value6, value7, value8, value9]

#Divide the value of each cell by total to get probabilities
for x in values:
    newvalues = x/totalnoofpatients
    probvalues.append(newvalues)
# Replace the values of the count matrix to the probability values
probability_matrix.replace(to_replace = [values], value = [probvalues], inplace = True)

# Use return statement to get the final result
probability_matrix 
```

```python
def confusionMatrixMetrics(confusionMatrix):
   """
   Calculate and return a Pandas DataFrame with various metrics from a given confusion matrix.

   Args:
    confusionMatrix (pandas.DataFrame): A confusion matrix of actual and predicted outcomes.
   """
   # Save the True/False Positive/Negative values from the input confusion matrix.
   TN = confusionMatrix.iloc[0][0]
   FN = confusionMatrix.iloc[1][0] 
   TP = confusionMatrix.iloc[1][1]
   FP = confusionMatrix.iloc[0][1]

   # Calculate various metrics.
   Population = TN+FN+TP+FP
   Prevalence = (TP+FP) / Population
   PPV = TP / (TP+FP)
   NPV = TN / (TN+FN)
   Sensitivity = TP / (TP+FN)
   Specificity = TN / (TN+FP) 
   FPR = FP / (TN+FP)
   FNR = FN / (TP+FN)

   # Create a new DataFrame storing to store the metrics.
   df_confmatrix = pd.DataFrame({'Metric':
                      ['TP', 'TN', 'FP', 'FN',
                       'Prevalence', 'PPV', 'NPV',
                       'Sensitivity', 'Specificity',
                       'FPR', 'FNR'],
                      'Value':
                      [TP, TN, FP, FN,
                       Prevalence, PPV, NPV,
                       Sensitivity, Specificity,
                       FPR, FNR]})
   df_confmatrix = df_confmatrix.set_index('Metric')
   
   return (df_confmatrix)
```

```python
confusionMatrixMetrics(ConfusionMatrix)
```
